```python
distance_posterior = np.array(GW_data["samples"]["distance"]) 
inclination_posterior = np.array(GW_data["samples"]["inclination"]) 
ra_posterior = np.array(GW_data["samples"]["ra"])* 180 / np.pi
dec_posterior = np.array(GW_data["samples"]["dec"])* 180 / np.pi
redshift_posterior = np.array(GW_data["samples"]["redshift"])
```

```python
#finding the statistics on ra and dec to find their counterparts in the galaxy catalog
mean_ra = np.mean(ra_posterior)
sigma_ra_down = np.quantile(ra_posterior, 15.7/100)
sigma_ra_up = np.quantile(ra_posterior, 1 - 15.7/100)
top = max(np.histogram(ra_posterior, density= True, bins =int(np.sqrt(len(ra_posterior))))[0])

plt.subplots(figsize = (12,8))
plt.vlines(sigma_ra_up, 0, top, label = "$1 \sigma$" , color = "green")
plt.vlines(sigma_ra_down, 0 ,top, label = "$1 \sigma$", color = "green")
plt.vlines(mean_ra, 0, top, label = "$ mean$", color = "red")
plt.hist(ra_posterior, histtype= "step", density=True, bins = int(np.sqrt(len(ra_posterior))) )
plt.title(gw_catalog[i]+" - RA posterior", pad = 20, fontsize = "xx-large")
plt.xlabel("RA degree", fontsize = "x-large")
plt.legend()


#############################################################################


mean_dec = np.mean(dec_posterior)
sigma_dec_down = np.quantile(dec_posterior, 15.7/100)
sigma_dec_up = np.quantile(dec_posterior, 1 - 15.7/100)
top = max(np.histogram(dec_posterior, density= True, bins =int(np.sqrt(len(dec_posterior))))[0])

plt.subplots(figsize = (12,8))
plt.vlines(sigma_dec_up, 0, top, label = "$1 \sigma$" , color = "green")
plt.vlines(sigma_dec_down, 0 ,top, label = "$1 \sigma$", color = "green")
plt.vlines(mean_dec, 0, top, label = "$ mean$", color = "red")
plt.hist(dec_posterior, histtype= "step", density=True, bins = int(np.sqrt(len(dec_posterior))) )
plt.title(gw_catalog[i]+" - DEC posterior", pad = 20, fontsize = "xx-large")
plt.xlabel("DEC degree", fontsize = "x-large")
plt.legend()

#############################################################################


mean_dist = np.mean(distance_posterior)
sigma_dist_down = np.quantile(distance_posterior, 15.7/100)
sigma_dist_up = np.quantile(distance_posterior, 1 - 15.7/100)
top = max(np.histogram(distance_posterior, density= True, bins =int(np.sqrt(len(distance_posterior))))[0])

plt.subplots(figsize = (12,8))
plt.vlines(sigma_dist_up, 0, top, label = "$1 \sigma$" , color = "green")
plt.vlines(sigma_dist_down, 0 ,top, label = "$1 \sigma$", color = "green")
plt.vlines(mean_dist, 0, top , label = "$ mean$", color = "red")
plt.hist(distance_posterior, histtype= "step", density=True, bins = int(np.sqrt(len(distance_posterior))) )
plt.title(gw_catalog[i]+" - $D_L$ posterior", pad = 20, fontsize = "xx-large")
plt.xlabel("$D_L$ $(Mpc)$", fontsize = "x-large")
plt.legend()


#############################################################################


mean_inc = np.mean(inclination_posterior)
sigma_inc_down = np.quantile(inclination_posterior, 15.7/100)
sigma_inc_up = np.quantile(inclination_posterior, 1 - 15.7/100)
top = max(np.histogram(inclination_posterior, density= True, bins =int(np.sqrt(len(inclination_posterior))))[0])

plt.subplots(figsize = (12,8))
plt.vlines(sigma_inc_up, 0, top, label = "$1 \sigma$ = "+str(np) , color = "green")
plt.vlines(sigma_inc_down, 0 ,top, label = "$1 \sigma$", color = "green")
plt.vlines(mean_inc, 0, top , label = "$ mean$", color = "red")
plt.hist(inclination_posterior, histtype= "step", density=True, bins = int(np.sqrt(len(inclination_posterior))) )
plt.title(gw_catalog[i]+" - Inclination angle posterior", pad = 20, fontsize = "xx-large")
plt.xlabel("$Inclination$ $(degree)$", fontsize = "x-large")
plt.legend()



#############################################################################


mean_z = np.mean(redshift_posterior)
sigma_z_down = np.quantile(redshift_posterior, 15.7/100)
sigma_z_up = np.quantile(redshift_posterior, 1 - 15.7/100)
top = max(np.histogram(redshift_posterior, density= True, bins =int(np.sqrt(len(redshift_posterior))))[0])


plt.subplots(figsize = (12,8))
plt.vlines(sigma_z_up, 0, top, label = "$1 \sigma$" , color = "green")
plt.vlines(sigma_z_down, 0 ,top, label = "$1 \sigma$", color = "green")
plt.vlines(mean_z, 0, top, label = "$ mean$", color = "red")
plt.hist(redshift_posterior, histtype= "step", density=True, bins = int(np.sqrt(len(redshift_posterior))) )
plt.title(gw_catalog[i]+" - redshift posterior", pad = 20, fontsize = "xx-large")
plt.xlabel("Redshift", fontsize = "x-large")
plt.legend()
```

```python
good_galaxies = (ra_galaxy >= sigma_ra_down) & (ra_galaxy <= sigma_ra_up) & (dec_galaxy >= sigma_dec_down) & (dec_galaxy <= sigma_dec_up) & (redshift_galaxy >= sigma_z_down) & (redshift_galaxy <= sigma_z_up)
    
good_galaxies_ra = ra_galaxy[good_galaxies]
good_galaxies_dec = dec_galaxy[good_galaxies]
good_galaxies_redshift = redshift_galaxy[good_galaxies]

len(good_galaxies_ra)
```
